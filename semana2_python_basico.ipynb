```python
# Números
edad = 50
print("Mi edad es:", edad)

# Texto
nombre = "Juan"
print("Me llamo", nombre)

# Booleano (verdadero/falso)
es_mayor = True
print("¿Soy mayor de edad?", es_mayor)

# Lista de números
numeros = [10, 20, 30, 40, 50]
print("Lista completa:", numeros)

# Primer y último elemento
print("Primer número:", numeros[0])
print("Último número:", numeros[-1])

# Suma de todos los números
print("Suma de la lista:", sum(numeros))

personas = [
    {'nombre': 'Ana', 'edad': 30},
    {'nombre': 'Juan', 'edad': 25}
]

# Busca el primer diccionario en la lista donde el 'nombre' es 'Ana' con un generador (iterador) y next
ana_dict = next((p for p in personas if p['nombre'] == 'Ana'), None)

# Si se encontró a Ana (es decir, ana_dict no es None)
if ana_dict:
    ana_dict['edad'] = 36/2

print(personas)

# ejemplo de For
numeros = [1, 2, 3, 4, 5]

for n in numeros:
    print("Número:", n)


# ejemplo de while
contador = 1
while contador <= 5:
    print("Contador:", contador)
    contador += 1
for n in range(1, 21):
    if n % 2 == 0:   # el operador % devuelve el resto
        print(n)
# Lista con mis comidas favoritas
comidas = ["pizza", "pasta", "ensalada"]
for a in comidas:
    print ("comida favorita:", a)
```
